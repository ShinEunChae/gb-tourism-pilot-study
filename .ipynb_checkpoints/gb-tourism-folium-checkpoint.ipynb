# 선정한 10개의 관광지 지도상에 시각화

```python
import pandas as pd
import folium
import base64
```

## 데이터프레임 생성

```python
spot_list = ['문경활공랜드', '문경 에코랄라', 'STX 문경 리조트', '카페 가은역', '오미나라',
             '안동시티투어 동이사람들', '구름에 리조트', '선성현문화단지한옥체험관', '카페 잇다', '주토피움']
```

```python
location_list = [[36.7362634516241, 128.15384156174355], [36.654107237437735, 128.06036232477607],
                 [36.56931954861005, 127.96421263079597], [36.649928168816565, 128.06039168779952],
                 [36.73468625016165, 128.09008152974548], [36.57434825672118, 128.67587414880057],
                 [36.57331067659036, 128.76871530727482], [36.697081543374864, 128.81311332606268],
                 [36.56449902334061, 128.7304819546278], [36.566741352358086, 128.78223178176773]]
```

```python
df = pd.DataFrame({'spot':spot_list, 'location':location_list})
```

```python
df
```

## 문경시
- 관광객들이 주로 이용할 문경시외버스터미널과 가장 유명한 관광지인 문경새재는 초록색 아이콘
- 선정한 관광지들은 파란색 아이콘으로 구별화

```python
m_mg = folium.Map([36.73338088007718, 128.107518755236], zoom_start=10)
folium.Marker([36.73338088007718, 128.107518755236], tooltip='문경시외버스터미널', icon=folium.Icon(icon='bookmark', color='green')).add_to(m_mg)
folium.Marker([36.7618765221891, 128.0766829806833], tooltip='문경새재', icon=folium.Icon(icon='bookmark', color='green')).add_to(m_mg)
```

```python
for i in range(5):
    pic = base64.b64encode(open('./img/%d.jpg'%(i+1), 'rb').read()).decode() # 팝업에 띄울 이미지 작업
    image_tag = '<img src="data:image/jpeg;base64,{}">'.format(pic)
    iframe = folium.IFrame(image_tag, width=310, height=200)
    popup = folium.Popup(iframe, max_width=650)
    folium.Marker(df['location'][i], popup=popup, tooltip='%s'%df['spot'][i], icon=folium.Icon(icon='star', color='blue')).add_to(m_mg)
```

```python
m_mg
```

## 안동시
- 관광객들이 주로 이용할 안동역은 초록색 아이콘
- 선정한 관광지들은 파란색 아이콘으로 구별화
- 시티투어버스 탑승장소를 빨간색 아이콘으로 추가

```python
m_ad = folium.Map([36.57435796045659, 128.67486337238125], zoom_start=11)
folium.Marker([36.57435796045659, 128.67486337238125], tooltip='안동역', icon=folium.Icon(icon='bookmark', color='green')).add_to(m_ad)
```

```python
for i in range(6, 10):
    pic = base64.b64encode(open('./img/%d.jpg'%(i+1), 'rb').read()).decode()
    image_tag = '<img src="data:image/jpeg;base64,{}">'.format(pic)
    iframe = folium.IFrame(image_tag, width=340, height=230)
    popup = folium.Popup(iframe, max_width=650)
    folium.Marker(df['location'][i], popup=popup, tooltip='%s'%df['spot'][i], icon=folium.Icon(icon='star', color='blue')).add_to(m_ad)
```

```python
m_ad
```

```python
# 투어버스 탑승장소 추가작업
busstop = [[36.5892957612231, 128.6105792072734],[36.574347942645666, 128.67589648424297],[36.56255977899123, 128.73279262419933]]
```

```python
for i in range(len(busstop)):
    folium.Marker(location=busstop[i], icon=folium.Icon(icon='check', color='red'), tooltip='투어버스탑승장소').add_to(m_ad)
```
