```python
from selenium import webdriver
from konlpy.tag import Okt
from nltk import Text
from matplotlib import font_manager, rc
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import time
```

```python
# selenium
driver = webdriver.Chrome(r'C:\Users\Shin\Desktop\assignment-gb\chromedriver.exe')
```

```python
url_list = []
content_list = ""
text = "문경%20|%20안동+관광" # '문경' 또는 '안동' 두 개 중 하나와 '관광' 키워드가 포함된 검색 결과
okt = Okt()
```

```python
# 불용어 사전
stopwords = '우리 생각 사람 것 이번 바로 그냥 조금 정도 듯 선택 매우 정말 엄청 그 외 후회 앞 별로 더 바보 이제 시작 항시 잠 못 잔 이번 후다닥 대해 제대로 도대체 와우 곳 포장 개 중 이 약 등 게 거 모습 제 수 날 때문 및 위해 안 때 도 하나 가족 공간 전 주문 내 다시 비롯 말 추천 간 위 타고 분 발 리 시간 명 중간 막 도착 왜 소요 제일 길 점 아이 예약 위치 출발 이야기 이상 코스 세트 이유 인터넷 훅 술 결정 다음 가기 해도 거의 종류 안주 꼭 하니 통과 참고 소개 테이블 밥 뭘 식사 몇 입 손 뒤 가요 나름 포스팅 검색 짐 난 또 고민 쪽 소스 상황 좀 나 진짜 요 옆 오늘 가장 비 볼 느낌 저 준비 한번 속 여기 보고 마음 사실 를 계획 주변 의 이름 단지 번 먼저 설명 다른 지금 더욱 직접 요즘 가격 주말 저희 모두 걸 아주 관련 할인 편 하루 근처 아래 잠시 현재 통해 침 처 최초 인기 제공 미리 방 확인 음식 오픈 택시 버스 통해 밑 중단 데 날씨 매력 엄마 또한 이동 정보 맛 후 여러분 건 사용 내부 가을 역시 마지막 모든 만 이용 방문 자리 문경시 안동시 대구'.split()
```

```python
for i in range(1, 501):  # 검색 결과의 500페이지까지 블로그 본문 크롤링
    url = 'https://section.blog.naver.com/Search/Post.nhn?pageNo='+ str(i) + '&rangeType=ALL&orderBy=sim&keyword=' + text # url 값 설정
    driver.get(url)
    time.sleep(0.5) # 오류를 방지하기 위해 0.5초의 간격을 둚
    
    for j in range(1, 3):
        titles = driver.find_element_by_xpath('/html/body/ui-view/div/main/div/div/section/div[2]/div['+ str(j) +']/div/div[1]/div[1]/a[1]')
        title = titles.get_attribute('href')
        url_list.append(title)

for url in url_list: # url_list에 저장된 블로그 주소를 하나씩 순회
    driver.get(url)
    driver.switch_to.frame('mainFrame')
    overlays = ".se-component.se-text.se-l-default"
    contents = driver.find_elements_by_css_selector(overlays)
    
    for content in contents:
        content_list = content_list + content.text # content_list에 블로그 본문을 차례로 누적함

myList = okt.pos(content_list, norm=True, stem=True) # 형태소 추출
myList_filter = [x for x, y in myList if y in ['Noun']] # 추출된 형태소 중 명사만 따로 저장
myList_sw = [i for i in myList_filter if i not in stopwords] # 불용어 제거

Okt = Text(myList_sw, name="Okt")

# 그래프에 한글이 깨지지 않고 출력되도록 수정
font_location = "c:/Windows/Fonts/malgun.ttf"
font_name = font_manager.FontProperties(fname=font_location).get_name()
rc('font', family=font_name)
plt.rcParams["figure.figsize"] = (20,16)

# 그래프의 x, y label 설정
plt.xlabel("명사")
plt.ylabel("빈도")

# 그래프의 x, y 값을 설정
wordInfo = dict()
for tags, counts in Okt.vocab().most_common(50):
    if(len(str(tags)) > 1):
        wordInfo[tags] = counts

values = sorted(wordInfo.values(), reverse=True)
keys = sorted(wordInfo, key=wordInfo.get, reverse=True)

plt.bar(range(len(wordInfo)), values, align='center')
plt.xticks(range(len(wordInfo)), list(keys), rotation='70')
plt.show()
 
# wordCloud 출력
wc = WordCloud(width = 2500, height = 1400, background_color="white", font_path=font_location, max_words=200)
plt.figure(figsize=(25, 14))
plt.imshow(wc.generate_from_frequencies(Okt.vocab()))
plt.axis("off")
plt.show()
```

```python
text = '안동+카페+잇다'

for i in range(1, 11):  # 검색 결과의 5페이지까지 블로그 본문 크롤링
    url = 'https://section.blog.naver.com/Search/Post.nhn?pageNo='+ str(i) + '&rangeType=ALL&orderBy=sim&keyword=' + text # url 값 설정
    driver.get(url)
    time.sleep(0.5) # 오류를 방지하기 위해 0.5초의 간격을 둚
    
    for j in range(1, 3):
        titles = driver.find_element_by_xpath('/html/body/ui-view/div/main/div/div/section/div[2]/div['+ str(j) +']/div/div[1]/div[1]/a[1]')
        title = titles.get_attribute('href')
        url_list.append(title)

for url in url_list: # url_list에 저장된 블로그 주소를 하나씩 순회
    driver.get(url)
    driver.switch_to.frame('mainFrame')
    overlays = ".se-component.se-text.se-l-default"
    contents = driver.find_elements_by_css_selector(overlays)
    
    for content in contents:
        content_list = content_list + content.text # content_list에 블로그 본문을 차례로 누적함

myList = okt.pos(content_list, norm=True, stem=True) # 형태소 추출
myList_filter = [x for x, y in myList if y in ['Noun'] or y in ['Verb'] or y in ['Adjective']] # 명사, 동사, 형용사 추출
myList_sw = [i for i in myList_filter if i not in stopwords] # 불용어 제거

Okt = Text(myList_sw, name="Okt")

# 그래프에 한글이 깨지지 않고 출력되도록 수정
font_location = "c:/Windows/Fonts/malgun.ttf"
font_name = font_manager.FontProperties(fname=font_location).get_name()
rc('font', family=font_name)
plt.rcParams["figure.figsize"] = (20,16)
 
# wordCloud 출력
wc = WordCloud(width = 2500, height = 1400, background_color="white", font_path=font_location, max_words=50)
plt.figure(figsize=(25, 14))
plt.imshow(wc.generate_from_frequencies(Okt.vocab()))
plt.axis("off")
plt.show()
```
